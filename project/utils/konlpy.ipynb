```python
import konlpy
import pandas as pd
from konlpy.tag import Okt
```

```python
df = pd.read_csv('../data/korean.csv')
```

```python
okt = Okt()
```

```python
okt.tagset.keys()
```

```python
tag_dict = {}

for k  in okt.tagset.keys():
    tag_dict[k] = []

## tagset.keys 에 없는 것들 추가
tag_dict['Modifier'] = []
tag_dict['VerbPrefix'] = []
```

```python
for s in df['text']:
    for t in okt.pos(s):
        tag_dict[t[1]].append(t[0])
        
        
df = pd.DataFrame.from_dict(tag_dict,orient='index')
df
```
